```python
# concat images
import cv2
from tqdm import tqdm
from google.colab.patches import cv2_imshow
```

```python
all_ds = ['train_orig_100', "train_data_scaling", "train_data_randCrop", "train_data_distortion", "train_data_blur", "train_data_erasing"]

images_to_concat = {}
for ds in all_ds:
  images_to_concat[ds] = []
  # iterate all 5 files and concat images
  for i in tqdm(range(5)):
    img = cv2.imread(f'/content/gdrive/My Drive/Research/saliency_images_overlap/' + ds + '/' + ds + '_' + str(i) + ".png")
    images_to_concat[ds].append(img)
```

```python
# concat 5 examples from each CNN vertically
images_vert = []
for ds in all_ds:
  images_vert.append(cv2.vconcat(images_to_concat[ds]))

# concat the 6 verticle strips horizontally
final_img = cv2.hconcat(images_vert)
cv2_imshow(final_img)

# save the output image
cv2.imwrite(f'/content/gdrive/My Drive/Research/saliency_overlap_concat.jpeg', final_img)
```
